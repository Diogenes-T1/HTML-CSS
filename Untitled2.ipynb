```python
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive, files

# Carregando o arquivo CSV
df = pd.read_csv("/content/drive/MyDrive/Atividade de extensão uninter/Stores.csv")

# Renomeando as colunas
df = df.rename(columns={"Store ID ": "ID",
                        "Store_Area": "Patio",
                        "Daily_Customer_Count": "Visitantes",
                        "Items_Available": "Itens Disponiveis",
                        "Store_Sales": "Vendas"})

# Cálculo dos valores mínimo, máximo, médio e desvio padrão
itens_min = df["Itens Disponiveis"].min()
itens_max = df["Itens Disponiveis"].max()
itens_mean = df["Itens Disponiveis"].mean()
itens_std = df["Itens Disponiveis"].std()

visitantes_min = df["Visitantes"].min()
visitantes_max = df["Visitantes"].max()
visitantes_mean = df["Visitantes"].mean()
visitantes_std = df["Visitantes"].std()

vendas_min = df["Vendas"].min()
vendas_max = df["Vendas"].max()
vendas_mean = df["Vendas"].mean()
vendas_std = df["Vendas"].std()

# Exibindo os valores mínimo, máximo, médio e desvio padrão
print("Valores para a coluna 'Itens':")
print(f"Mínimo: {itens_min}")
print(f"Máximo: {itens_max}")
print(f"Média: {itens_mean}")
print(f"Desvio Padrão: {itens_std}")
print()

print("Valores para a coluna 'Visitantes':")
print(f"Mínimo: {visitantes_min}")
print(f"Máximo: {visitantes_max}")
print(f"Média: {visitantes_mean}")
print(f"Desvio Padrão: {visitantes_std}")
print()

print("Valores para a coluna 'Vendas':")
print(f"Mínimo: {vendas_min}")
print(f"Máximo: {vendas_max}")
print(f"Média: {vendas_mean}")
print(f"Desvio Padrão: {vendas_std}")
print()

# Definindo as cores
colors = ['FireBrick', 'SteelBlue', 'DarkOrchid', 'black']

# Plotando o gráfico para a coluna "Itens"
plt.figure(figsize=(7, 5))
plt.bar(["Mínimo", "Máximo", "Média", "Desvio Padrão"], [itens_min, itens_max, itens_mean, itens_std], color=colors)
plt.xlabel("Estatísticas")
plt.ylabel("Itens Disponíveis")
plt.title("Estatísticas de Itens Disponíveis")
plt.grid(True)
plt.show()

# Plotando o gráfico para a coluna "Visitantes"
plt.figure(figsize=(7, 5))
plt.bar(["Mínimo", "Máximo", "Média", "Desvio Padrão"], [visitantes_min, visitantes_max, visitantes_mean, visitantes_std], color=colors)
plt.xlabel("Estatísticas")
plt.ylabel("Visitantes")
plt.title("Estatísticas de Visitantes")
plt.grid(True)
plt.show()

# Plotando o gráfico para a coluna "Vendas"
plt.figure(figsize=(7, 5))
plt.bar(["Mínimo", "Máximo", "Média", "Desvio Padrão"], [vendas_min, vendas_max, vendas_mean, vendas_std], color=colors)
plt.xlabel("Estatísticas")
plt.ylabel("Vendas")
plt.title("Estatísticas de Vendas")
plt.grid(True)
plt.show()
```
